```
import pandas as pd
from bsedata.bse import BSE  
import time
from pyspark.sql import SparkSession

pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)

github_excel_url = "https://github.com/anandaiml19/DE_Zoomcamp_Project2/raw/main/Equity%20(1).xlsx"
engine = 'openpyxl'  
NewEquityDF = pd.read_excel(github_excel_url, engine = engine)
NewEquityDF.head(n=5)
```

```
list_of_first50_StockIDs=["ABB",
"AEGISLOG",
"ARE&M",
"AMBALALSA",
"ANDHRAPET",
"ANSALAPI",
"UTIQUE",
"ARUNAHTEL",
"BOMDYEING",
"ASIANHOTNR",
"ATUL",
"ATVPR",
"BAJAJELEC",
"BAJAJHIND",
"FORCEMOT",
"BAJFINANCE",
"BALRAMCHIN",
"BANCOINDIA",
"CENTURYTEX",
"BANARISUG",
"BASF",
"BATAINDIA",
"BEML",
"BEL",
"BEPL",
"BIHSPONG",
"BINANIIND",
"BIRLACABLE",
"BLUESTARCO",
"DISAQ",
"QUESTCAP",
"BPL",
"OAL",
"CENTEXT",
"CESC",
"CHAMBLFERT",
"EXIDEIND",
"CIPLA",
"DICIND",
"CRISIL",
"CGPOWER",
"DABUR",
"DALMIASUG",
"ARVIND",
"BHEL",
"HINDPETRO",
"IFCI",
"MTNL",
"MRPL",
"CHENNPETRO"
]

NewEquityDF['Security Code'] = NewEquityDF['Security Code'].astype(str) 
nifty50_OnlyDF= NewEquityDF[NewEquityDF['Security Id'].isin(list_of_first50_StockIDs)].reset_index(drop=True)
nifty50_OnlyDF.head(5)
```

```
nifty50_OnlyDF.columns = nifty50_OnlyDF.columns.str.replace(' ', '')
```

```
nifty50_OnlyDF
```

CALLING BSE API to fetch details about the 50 stocks 

```
bseObject = BSE(update_codes=True)  ##creating Bse Lib Object
listof_stockDicts = []
sqcode_ListNifty50 = nifty50_OnlyDF['SecurityCode'].values  
for sqCode in sqcode_ListNifty50:
    try:
        stockDict = bseObject.getQuote(sqCode)
        stockDict.pop("buy", None)
        stockDict.pop("sell", None)
        listof_stockDicts.append(stockDict)
        time.sleep(0.5)
    except IndexError:
        print(f"IndexError for {sqCode}: Data not available")

listof_stockDicts
```

```
newdf=pd.DataFrame(listof_stockDicts)
newdf.head(5)
```

TRANSFORMATION

```
newdf.rename(columns={'group': 'sharegroup'}, inplace=True)
newdf.rename(columns={'52weekHigh': 'fiftytwoweekHigh'}, inplace=True)
newdf.rename(columns={'52weekLow': 'fiftytwoweekLow'}, inplace=True)
newdf.rename(columns={'2WeekAvgQuantity': 'twoWeekAvgQuantity'}, inplace=True)
# Convert 'updatedOn' column to datetime and extract date
newdf['updatedOn'] = pd.to_datetime(newdf['updatedOn'], format='%d %b %y | %I:%M %p', errors='coerce')
```

```
newdf.head(3)
```

```
newdf=newdf.dropna()
```

```
newdf.isna().sum()
```

```
# Check if there are any invalid or missing date values
if pd.isna(newdf['updatedOn']).any():
    print("There are invalid or missing date values in the 'updatedOn' column.")
else:
    # Extract date from 'updatedOn' column and convert the column to datetime
    newdf['updatedOn'] = pd.to_datetime(newdf['updatedOn'].dt.date)
```

```
newdf.head(3)
```

```
if 'totalTradedValueCr' not in newdf.columns:
   # Assuming newdf is your DataFrame
    newdf['totalTradedValueCr'] = pd.to_numeric(newdf['totalTradedValue'].str.replace(',', '').str.replace(' Cr.', '', regex=True), errors='coerce')  # Convert to numeric and handle 'Cr.'
    newdf['totalTradedQuantityLakh'] = pd.to_numeric(newdf['totalTradedQuantity'].str.replace(',', '').str.replace(' Lakh', '', regex=True), errors='coerce')  # Convert to numeric and handle 'Lakh'
    newdf['twoWeekAvgQuantityLakh'] = pd.to_numeric(newdf['twoWeekAvgQuantity'].str.replace(',', '').str.replace(' Lakh', '', regex=True), errors='coerce')  # Convert to numeric and handle 'Lakh'
    newdf['marketCapFullCr'] = pd.to_numeric(newdf['marketCapFull'].str.replace(',', '').str.replace(' Cr.', '', regex=True), errors='coerce')  # Convert to numeric and handle 'Cr.'
    newdf['marketCapFreeFloatCr'] = pd.to_numeric(newdf['marketCapFreeFloat'].str.replace(',', '').str.replace(' Cr.', '', regex=True), errors='coerce')  # Convert to numeric and handle 'Cr.'

    # Drop original columns
    newdf.drop(['totalTradedValue', 'totalTradedQuantity','twoWeekAvgQuantity', 'marketCapFull', 'marketCapFreeFloat'], axis=1, inplace=True)
```
